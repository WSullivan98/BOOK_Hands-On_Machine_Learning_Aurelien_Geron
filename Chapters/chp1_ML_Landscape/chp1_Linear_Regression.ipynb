```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
import numpy as np
import pandas as pd
import sklearn.linear_model
```

# Load the data

```python
oecd_bli = pd.read_csv('oecd_bli_2015_2021.csv', thousands=',')
gdp_per_capita = pd.read_csv('gdp_per_capita.csv', thousands=',', encoding='latin1', na_values='n/a')
```

```python
oecd_bli.head()
```

```python
gdp_per_capita = gdp_per_capita.rename(columns={'ï»¿Country':'Country'}, inplace=False)
gdp_per_capita.head()
```

# Prep the data

```python
def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd_bli = oecd_bli[oecd_bli['INEQUALITY']=='TOT']
    oecd_bli = oecd_bli.pivot(index='Country', columns='Indicator', values="Value")
    
    gdp_per_capita.rename(columns={'2015': 'GDP per capita'}, inplace=True)
#    gdp_per_capita.set_index('Country', inplace=True)
    
    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                                 left_index=True, right_index=True)
    full_country_stats.sort_values(by='GDP per capita', inplace=True)
#    remove_indices = [0,1,6,8,33,34,35]
    return full_country_stats[['GDP per capita', 'Life satisfaction']]
```

```python
full_country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)
full_country_stats
```

```python
country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)
X = np.c_[country_stats['GDP per capita']]
Y = np.c_[country_stats['Life satisfaction']]
```

```python
X
```

```python
Y
```

# Visualize the data

```python
country_stats.plot(kind='scatter', x='GDP per capita', y='Life satisfaction',figsize=(12,8))
plt.show()
```

# Select a linear model

```python
model = sklearn.linear_model.LinearRegression()
```

# Train the model

```python
model.fit(X,Y)
```

# Make a prediction for USA

```python
X_new =  [[56848.51]]
print('Current US Life Satisfaction ', country_stats.loc['United States'],'\n', 'Predicted Life Satifaction ', model.predict(X_new))
```
